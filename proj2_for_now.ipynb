##### Lessons from Pulling API Data

Unfortunately, while pulling nba ticket data and getting other event ticket data I made two separate, consecutive calls back to back that duplicated most of the NBA entries. So I have duplicate data from almost every time I pulled data. The following code compares a list of sorted timestamps for every instance of a given event and if they are less than 10 minutes (600 "total_seconds") apart the index of the latter timestamp gets removed from the filtered dataframe (newdf = df[df.event_txt == g]).

```python
df = nba_tix
#lots of different dataframes to keep track of...almost done cleaning irrelevant data
df_tix = pd.DataFrame()
for g in nba_games:
    newdf = df[df.event_txt == g].sort_index()
    count_rows = newdf.count()[0]
    id_list = []
    for idx in range(0, (count_rows - 1)):
        a = newdf.iloc[idx].datetime
        b = newdf.iloc[idx + 1].datetime
        delta = (abs(b-a))
        if delta.total_seconds() < 600:
            id_list.append(idx)
    newdf = newdf.drop(newdf.index[id_list])
    if len(df_tix) == 0:
        df_tix = newdf.copy()
    else:
        df_tix = pd.concat([newdf, df_tix])

print len(df_tix)
```

##### Target Variable

I am guessing that when better teams are playing that the tickets are going to be more volatile and will generally be more expensive. I am going to see if I can predict whether the home team is playing a team ranked in the top, middle or bottom of their respective conference(eastern/western) based on the total change in average ticket price multiplied by the mean of the average ticket price. I pulled the rankings from nba.com.

```python
nba_rating = {'celtics' : 'med', 'nets': 'low', 'knicks' : 'low', '76ers' : 'low', 'raptors' : 'high',
	'mavericks' : 'med', 'rockets': 'high', 'grizzlies': 'high', 'pelicans' : 'med', 'spurs' : 'med',
	'bulls' : 'high', 'cavaliers' : 'high', 'pistons': 'low', 'pacers' : 'med', 'bucks' : 'med', 'nuggets' : 'low',
	'timberwolves' : 'low', 'thunder' : 'med', 'blazers' : 'high', 'jazz' : 'low',
	'hawks' : 'high', 'hornets' : 'med', 'heat' : 'med', 'magic' : 'low','wizards' : 'high',
	'warriors' : 'high', 'clippers' : 'high', 'lakers' : 'low', 'suns' : 'med', 'kings' : 'low'}

df_tix['opp_ranking'] = 0

#It might be useful to have the home team's ranking as well to see if that helps prediction
df_tix['home_ranking'] = 0

for g in nba_games:
    t = g.lower()
    t = t.split('_')
    #away team always listed first
    opp = t[0]
    home = t[1]
    df_tix.ix[df_tix.event_txt == g, 'opp_ranking'] = nba_rating.get(opp)
    df_tix.ix[df_tix.event_txt == g, 'home_ranking'] = nba_rating.get(home)

df_tix.head(25)
```

```python
df_tix['delta_avg'] = 0
#guessing that games with better teams will have more action, volitility
#avg.max - avg.min * avg.mean
for g in nba_games:
    snewdf = df_tix[df_tix.event_txt == g]
    t = snewdf.avg
    delta = t.max() - t.min()
    df_tix.ix[df_tix.event_txt == g, 'delta_avg'] = delta

df_tix.describe()
```
