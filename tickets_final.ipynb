```python
#this is the csv that contains the data from my api calls
url = 'https://raw.githubusercontent.com/hpredieri/gassign/master/nbatickets.csv'
cols = ['event_id', 'venue_id', 'event_txt', 'listings', 'avg', 'low', 'high', 'datetime']
nba_data = pd.read_csv(url, names=cols, parse_dates=['datetime'], dayfirst=True)
nba_data.tail()
```

```python
#I manually changed the event_txt when I made the api call so if there are two 'teams' then it has an underscore
def ballgame(row):
    if '_' in row['event_txt']:
        return 1
    else:
        return 0
    
nba_data['ballgame'] = nba_data.apply(ballgame, axis=1)

#to get dates for a timeseries. Unfortunately only 6 days worth of data, so timeseries does not work too well
newlist = []
for i in nba_data.index:
    new=nba_data.datetime[i].date()
    newlist.append(new)

nba_data['date']=newlist
nba_data.head()

nba_data = nba_data.set_index(nba_data.datetime)

sports_events = nba_data[nba_data.ballgame==1]
non_sports = nba_data[nba_data.ballgame==0]
```

```python
#finding events whose listings and average price have the greatest standard deviation
#will be respectively evlist_dev and evavg_dev

evlist_dev = by_event.listings.std().dropna()
top_eld = evlist_dev.nlargest(n=15)

evavg_dev = by_event.avg.std().dropna()
top_ead = evavg_dev.nlargest(n=15)

sports_gp = sports_events.groupby('event_id')
sports_dev = sports_gp.avg.std().dropna()
top_sports = sports_dev.nlargest(n=15)

nonsports_gp = non_sports.groupby('event_id')
ns_dev = nonsports_gp.avg.std().dropna()
top_ns = ns_dev.nlargest(n=15)
```

```python
#finding the most active events and making a filter for the original data_set
active_events = list()
top_lists = [top_eld, top_ead, top_sports, top_ns]
for lst in top_lists:
    x = lst.iteritems()
    for k, v in x:
        if k not in active_events:
            active_events.append(k)

print len(active_events)
```

```python
active = nba_data[nba_data.event_id.isin(active_events) ==True]
print len(active)
```
