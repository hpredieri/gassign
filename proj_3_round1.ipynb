```python
cols = ['event_id', 'venue_id', 'event_txt', 'listings', 'avg', 'low', 'high', 'event_date','datetime']
tick_data = 'C:/Users/owner/Documents/GAdatascience/scalpyr/scalpyr/nbatickets23.csv'
nba_data = pd.read_csv(tick_data, names=cols, parse_dates=['datetime', 'event_date'], dayfirst=True)
```

```python
nba_data.info()
tickets = nba_data.dropna()
tickets.info()
```

```python
#Here is the list of all the teams in the nba
nba_slugs = ['boston-celtics', 'brooklyn-nets', 'new-york-knicks', 'philadelphia-76ers', 'toronto-raptors',
	'dallas-mavericks', 'houston-rockets', 'memphis-grizzlies', 'new-orleans-pelicans', 'san-antonio-spurs',
	'chicago-bulls', 'cleveland-cavaliers', 'detroit-pistons', 'indiana-pacers', 'milwaukee-bucks', 'denver-nuggets',
	'minnesota-timberwolves', 'oklahoma-city-thunder', 'portland-trail-blazers', 'utah-jazz',
	'atlanta-hawks', 'charlotte-hornets', 'miami-heat', 'orlando-magic','washington-wizards',
	'golden-state-warriors', 'los-angeles-clippers', 'los-angeles-lakers', 'phoenix-suns', 'sacromento-kings']
nba_teams = list()
for t in nba_slugs:
    t = t.split('-')
    #the mascot is always the last listed last
    t = t[-1]
    nba_teams.append(t)
```

```python
#construct new list with only nba games by finding events with nba team names
unique_events = tickets.event_txt.unique()
print len(unique_events)

nba_games = list()
for u in unique_events:
    #all nba games have an underscore in the event_txt so it 'finds' where underscores do not exist
    if u.find('_') == -1:
        continue
    else:
        t = u.split('_')
        t1 = t[0]
        t1 = t1.lower()
        t2 = t[1]
        t2 = t2.lower()
        if t1 in nba_teams:
            #I found some NHL teams share names with NBA teams and were not filtered out so I need to check both team names
            if t2 in nba_teams:
                nba_games.append(u)
            else:
                continue
        else:
            continue
            
print len(nba_games)
#construct new df with only nba games (345) 
```

```python
#construct new df with only nba games (345) events
nba_tix = pd.DataFrame()
for g in nba_games:
    newdf = tickets[tickets.event_txt == g]
    if len(nba_tix) == 0:
        nba_tix = newdf.copy()
    else:
        nba_tix = pd.concat([nba_tix, newdf])
print len(nba_tix)
```

```python
df = nba_tix
#lots of different dataframes to keep track of...almost done cleaning irrelevant data
df_tix = pd.DataFrame()
for g in nba_games:
    newdf = df[df.event_txt == g].sort_index()
    count_rows = newdf.count()[0]
    id_list = []
    for idx in range(0, (count_rows - 1)):
        a = newdf.iloc[idx].datetime
        b = newdf.iloc[idx + 1].datetime
        delta = (abs(b-a))
        if delta.total_seconds() < 600:
            id_list.append(idx)
    newdf = newdf.drop(newdf.index[id_list])
    if len(df_tix) == 0:
        df_tix = newdf.copy()
    else:
        df_tix = pd.concat([newdf, df_tix])

print len(df_tix)
```
