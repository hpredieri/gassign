```python
#see if there is a big difference between atemp and temp
bike_data['dtemp'] = abs(bike_data.atemp - bike_data.temp)
print bike_data.dtemp.mean()
print bike_data.dtemp.max()
print bike_data.dtemp.min()


#scatter matrix to look for potentially interesting patterns
cols = ['windspeed', 'temp','hum','registered', 'casual']
pd.scatter_matrix(bike_data[cols], figsize=(12, 12))
```

Looking at this matrix you can find patterns that would agree with some common assumptions that casual riders are much more averse to inclimate weather conditions like high wind speeds and low temperatures.

It seems that the registered riders follow those general patterns but have more outliers(dates) on the extremes of both windspeed and temperature.

```python
#get a closer look at the lower temperatures to find any odd distributions
low_temps = bike_data[bike_data.temp <= .25]
#understand number of instances
print "Number of instances :", len(low_temps)
print "Percent of total instances:", (float(len(low_temps)) /len(bike_data))

#check daily and hourly for registered riders
plt.figure()
low_temps.groupby(['weekday']).registered.sum().plot(kind='bar')

plt.figure()
low_temps.groupby(['hr']).registered.sum().plot(kind='bar')

#compare to full data set for registered riders
plt.figure()
bike_data.groupby(['weekday']).registered.sum().plot(kind='bar')

plt.figure()
bike_data.groupby(['hr']).registered.sum().plot(kind='bar')
```

From the daily graphs we can see that some of those registered riders are psychologically affected by the combined difficulty of the cold weather and "hump day" and take it as an excuse to find another route of transportation.

An interesting phenomenon we see here is that compared to the normal bimodal distribution that we saw in the original graph where registered riders are coming and going at 8 AM and 5 PM, when the weather is relatively cold they are much less more unlikely to ride back at 5pm.
